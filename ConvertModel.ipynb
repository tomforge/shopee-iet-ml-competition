```python
import numpy as np
from keras.models import load_model, Model
from keras.preprocessing.image import ImageDataGenerator
from keras.applications import inception_v3
from keras.layers import Dense, Dropout
import pandas as pd
from src.utils import printModelMetric, analyze
```

```python
model1 = inception_v4.InceptionV4(include_top=False, input_shape=(256,256,3), pooling='avg')
```

```python
TEST_DIR = 'data/test'
IMG_WIDTH, IMG_HEIGHT = 256, 256
NUM_PREDICT = 3813

x = model1.output
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
preds = Dense(18, activation='softmax')(x)

model = Model(inputs=model1.input, outputs=preds)
        
```

```python
model.load_weights("inceptionv3_85_hinge.h5")
```

```python
# Need to apply the same data preprocessing
test_datagen = ImageDataGenerator(
    rescale = 1./255)

# Set no shuffle so the predictions follow file order
test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size = (IMG_HEIGHT, IMG_WIDTH),
    class_mode = "categorical",
    shuffle = False)
```

```python
results = model.predict_generator(test_generator, verbose=True)
```

```python
pred_labels = np.argmax(results, axis=1)
printModelMetric(analyze(pred_labels), "inception_v3_yuj_hinge")
```
